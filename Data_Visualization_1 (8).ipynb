```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
df = sns.load_dataset('titanic')

# Display the first few rows
df.head()
```

```python
sns.countplot(x='sex', hue='survived', data=df, palette='Set2')
plt.title('Survival Count Based on Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='pclass', hue='survived', data=df, palette='Set1')
plt.title('Survival Count Based on Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
sns.histplot(df['age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
sns.histplot(df['fare'], bins=30, kde=True, color='salmon')
plt.title('Ticket Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
print("""
- Most passengers paid a lower fare (under 100), but a few outliers paid very high ticket prices.
- Higher-class passengers (1st class) generally had higher fares and better survival chances.
""")
```

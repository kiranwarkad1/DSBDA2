```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Iris dataset
iris = sns.load_dataset('iris')

# Show first 5 rows
iris.head()
```

```python
# Print features and types
print(iris.dtypes)
```

```python
# Histogram
plt.hist(iris['sepal_length'], bins=20, color='skyblue')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()

# Boxplot
sns.boxplot(x=iris['sepal_length'], color='skyblue')
plt.title('Boxplot of Sepal Length')
plt.show()
```

```python
# Histogram
plt.hist(iris['sepal_width'], bins=20, color='lightgreen')
plt.title('Histogram of Sepal Width')
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
plt.show()

# Boxplot
sns.boxplot(x=iris['sepal_width'], color='lightgreen')
plt.title('Boxplot of Sepal Width')
plt.show()
```

```python
# Histogram
plt.hist(iris['petal_length'], bins=20, color='salmon')
plt.title('Histogram of Petal Length')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.show()

# Boxplot
sns.boxplot(x=iris['petal_length'], color='salmon')
plt.title('Boxplot of Petal Length')
plt.show()
```

```python
# Histogram
plt.hist(iris['petal_width'], bins=20, color='violet')
plt.title('Histogram of Petal Width')
plt.xlabel('Petal Width')
plt.ylabel('Frequency')
plt.show()

# Boxplot
sns.boxplot(x=iris['petal_width'], color='violet')
plt.title('Boxplot of Petal Width')
plt.show()
```

```python
print("""
Observations:
- Sepal Width shows a few outliers.
- Sepal Length, Petal Length, Petal Width distributions are mostly normal.
- Petal measurements are tighter and smaller in range compared to Sepal measurements.
""")
```

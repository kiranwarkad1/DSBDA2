```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
df = sns.load_dataset('titanic')

# Show first 5 rows
df.head()
```

```python
# Plot boxplot
plt.figure(figsize=(10,6))
sns.boxplot(x='sex', y='age', hue='survived', data=df, palette='Set2')
plt.title('Age Distribution by Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python
print("""
Observations:
- Females had higher survival rates across almost all age groups.
- Males had lower survival rates, especially in the 20-40 age group.
- Younger passengers (children) had better survival chances compared to older passengers.
- Outliers are visible, especially among older ages.
""")
```
